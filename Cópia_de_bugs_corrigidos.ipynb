```python
import pandas as pd
import matplotlib.pyplot as plt

# Criando o DataFrame com os dados fornecidos
dados = {
    'Dia da Semana': ['Segunda-feira', 'Terça-feira', 'Quarta-feira', 'Quinta-feira', 'Sexta-feira', 'Sábado', 'Domingo'],
    'Horas Trabalhadas': [6, 7, 8, 6, 7, 5, 4],
    'Bugs Corrigidos': [3, 2, 1, 4, 3, 2, 1],
    'Tarefas Concluídas': [5, 4, 6, 4, 5, 3, 2]
}

df = pd.DataFrame(dados)

# Gráfico de barras para Horas Trabalhadas, Bugs Corrigidos e Tarefas Concluídas
plt.figure(figsize=(10, 6))
bar_width = 0.2
bar_positions = list(range(len(df)))

plt.bar(bar_positions, df['Horas Trabalhadas'], width=bar_width, label='Horas Trabalhadas', color='#1f77b4')
plt.bar([p + bar_width for p in bar_positions], df['Bugs Corrigidos'], width=bar_width, label='Bugs Corrigidos', color='#2ca02c')
plt.bar([p + 2*bar_width for p in bar_positions], df['Tarefas Concluídas'], width=bar_width, label='Tarefas Concluídas', color='#d62728')

plt.xticks([p + bar_width for p in bar_positions], df['Dia da Semana'])
plt.xlabel('Dia da Semana')
plt.ylabel('Quantidade')
plt.title('Progresso Diário - Horas Trabalhadas, Bugs Corrigidos e Tarefas Concluídas')
plt.legend()
plt.show()
```

#<font color=violet> Média de Bugs Corrigidos

```python
import pandas as pd
import matplotlib.pyplot as plt

# Criando o DataFrame com os dados fornecidos
dados = {
    'Dia da Semana': ['Segunda-feira', 'Terça-feira', 'Quarta-feira', 'Quinta-feira', 'Sexta-feira', 'Sábado', 'Domingo'],
    'Horas Trabalhadas': [6, 7, 8, 6, 7, 5, 4],
    'Bugs Corrigidos': [3, 2, 1, 4, 3, 2, 1],
    'Tarefas Concluídas': [5, 4, 6, 4, 5, 3, 2]
}

df = pd.DataFrame(dados)

# Calculando a média de bugs corrigidos
media_bugs_corrigidos = df['Bugs Corrigidos'].mean()

# Gráfico de barras para Horas Trabalhadas, Bugs Corrigidos e Tarefas Concluídas
plt.figure(figsize=(10, 6))
bar_width = 0.2
bar_positions = list(range(len(df)))

plt.bar(bar_positions, df['Horas Trabalhadas'], width=bar_width, label='Horas Trabalhadas', color='#1f77b4')
plt.bar([p + bar_width for p in bar_positions], df['Bugs Corrigidos'], width=bar_width, label='Bugs Corrigidos', color='#2ca02c')
plt.bar([p + 2*bar_width for p in bar_positions], df['Tarefas Concluídas'], width=bar_width, label='Tarefas Concluídas', color='#d62728')

plt.xticks([p + bar_width for p in bar_positions], df['Dia da Semana'])
plt.xlabel('Dia da Semana')
plt.ylabel('Quantidade')
plt.title('Progresso Diário - Horas Trabalhadas, Bugs Corrigidos e Tarefas Concluídas')
plt.legend()
plt.show()

# Gerando o relatório com a média de bugs corrigidos
relatorio = f"RELATÓRIO DE PROGRESSO DIÁRIO - DESENVOLVIMENTO DE SOFTWARE\n"
relatorio += f"Média de Bugs Corrigidos: {media_bugs_corrigidos:.2f} bugs por dia\n\n"
relatorio += f"Período: {df['Dia da Semana'].iloc[0]} a {df['Dia da Semana'].iloc[-1]}\n\n"
relatorio += f"1. Total de Horas Trabalhadas: {df['Horas Trabalhadas'].sum()} horas\n"
relatorio += f"2. Média Diária de Horas Trabalhadas: {df['Horas Trabalhadas'].mean():.2f} horas\n"
relatorio += f"3. Total de Bugs Corrigidos: {df['Bugs Corrigidos'].sum()} bugs\n"
relatorio += f"4. Média Diária de Bugs Corrigidos: {df['Bugs Corrigidos'].mean():.2f} bugs por dia\n"
relatorio += f"5. Total de Tarefas Concluídas: {df['Tarefas Concluídas'].sum()} tarefas\n"
relatorio += f"6. Média Diária de Tarefas Concluídas: {df['Tarefas Concluídas'].mean():.2f} tarefas por dia\n"

print(relatorio)
```

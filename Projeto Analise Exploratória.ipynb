#<font color=violet>Análise Exploratória

 - Vamos realizar uma análise exploratória básica dos dados fornecidos para obter mais insights sobre o progresso diário do projeto de desenvolvimento de
software.
Primeiro, vamos criar o DataFrame com os dados fornecidos e realizar algumas visualizações gráficas para entender melhor as tendências ao longo da semana.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Criando o DataFrame com os dados fornecidos
dados = {
    'Dia da Semana': ['Segunda-feira', 'Terça-feira', 'Quarta-feira', 'Quinta-feira', 'Sexta-feira', 'Sábado', 'Domingo'],
    'Horas Trabalhadas': [6, 7, 8, 6, 7, 5, 4],
    'Bugs Corrigidos': [3, 2, 1, 4, 3, 2, 1],
    'Tarefas Concluídas': [5, 4, 6, 4, 5, 3, 2]
}

df = pd.DataFrame(dados)

# Visualizando os dados
print(df)
```

- Agora, vamos criar gráficos para visualizar as métricas ao longo da semana:

```python
# Gráfico de barras para Horas Trabalhadas, Bugs Corrigidos e Tarefas Concluídas
plt.figure(figsize=(10, 6))
plt.bar(df['Dia da Semana'], df['Horas Trabalhadas'], label='Horas Trabalhadas', color='b')
plt.bar(df['Dia da Semana'], df['Bugs Corrigidos'], label='Bugs Corrigidos', color='g')
plt.bar(df['Dia da Semana'], df['Tarefas Concluídas'], label='Tarefas Concluídas', color='r')
plt.xlabel('Dia da Semana')
plt.ylabel('Quantidade')
plt.title('Progresso Diário - Horas Trabalhadas, Bugs Corrigidos e Tarefas Concluídas')
plt.legend()
plt.show()
```

- O gráfico de barras acima mostra a quantidade de horas trabalhadas, bugs corrigidos e tarefas concluídas em cada dia da semana. Isso nos ajuda a entender a distribuição do progresso diário em cada uma das métricas.

- Agora, vamos criar um gráfico de linha para visualizar a produtividade diária:

```python
# Calculando a produtividade diária (Tarefas Concluídas por Hora)
df['Produtividade Diária'] = df['Tarefas Concluídas'] / df['Horas Trabalhadas']

# Gráfico de linha para Produtividade Diária
plt.figure(figsize=(10, 6))
plt.plot(df['Dia da Semana'], df['Produtividade Diária'], marker='o', color='purple')
plt.xlabel('Dia da Semana')
plt.ylabel('Produtividade Diária (Tarefas Concluídas por Hora)')
plt.title('Produtividade Diária')
plt.grid()
plt.show()
```

- O gráfico de linha acima mostra a produtividade diária, medida pelo número de tarefas concluídas por hora de trabalho, em cada dia da semana.
